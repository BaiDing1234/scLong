```python
import pickle
from sklearn import metrics
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('fpr_tpr_scfm_run2.pkl','rb') as f:
    fpr,tpr=pickle.load(f)
with open('fpr_tpr_bsl_run4.pkl','rb') as f:
    fpr_b,tpr_b=pickle.load(f)
```

```python
fig = plt.figure(figsize=(5,5),dpi=200)
ax = plt.subplot(111)
ax.plot(fpr,tpr,c='r',label='ScFM',lw=2.5)
ax.plot(fpr_b,tpr_b,c='g',label='DeepDDS',lw=2.5)
ax.plot(np.arange(0,1,1/100),np.arange(0,1,1/100),c='grey',ls='--')
ax.spines[['right','top']].set_visible(False)
for axis in ['bottom','left']:
    ax.spines[axis].set_linewidth(3)
ax.set_xlabel('FPR',fontdict={'size':15})
ax.set_ylabel('TPR',fontdict={'size':15})
ax.legend()
ax.set_title('ROC Curve',fontdict={'size':20})
```
