```python
import numpy as np
import pandas as pd
import pickle
```

```python

gene2vec_scgpt=np.load('/data3/ruiyi/scgpt/gene2vec_scgpt.npy')
gene2vec_scgpt.shape
```

```python
split='train'
df=pd.read_csv('../signature_{}.csv'.format(split))

import json

# Path to the JSON file
json_file_path = "/data3/ruiyi/scgpt/save/vocab.json"

# Open the JSON file and load the data into a Python dictionary
with open(json_file_path, 'r') as file:
    name2idx = json.load(file)


idx2idx={}
for k,name in enumerate(df.keys()):
    if name in name2idx:
        print(name)
        idx=name2idx[name]
        idx2idx[idx]=k-5

```

```python
cnt=0
deepce_scgpt_embed=np.zeros((len(df.keys())-5,gene2vec_scgpt.shape[-1]))
for i in range(gene2vec_scgpt.shape[0]):
    if i in idx2idx:
        cnt+=1
        deepce_scgpt_embed[idx2idx[i]]=gene2vec_scgpt[i]
deepce_scgpt_embed.shape,cnt
```

```python

np.save('/data1/ruiyi/deepce/deepce_scgpt_gene2vec_embed.npy',deepce_scgpt_embed)
```
